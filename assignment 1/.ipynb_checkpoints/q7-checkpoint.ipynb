```python
filenames = ["A01", "A02", "A03","A04", "A05", "A06", "A07", "A08", "A09", "A10", "A11", "A12", "A13", "A14", "A19", "A20"]
```

```python
hashmap = {}
total = 0
for filename in filenames:
    file = open("fc2/"+ filename)
    for i in file:
        j = i.split()
        if j[2] not in hashmap:
            hashmap[j[2]] = 0
        hashmap[j[2]] += 1  
        total += 1
    
print(hashmap)
```

```python
total
```

```python
distri = []
for i in hashmap:
    distri.append(hashmap[i]/total)
```

```python
distri
```

```python
import math

def entropy(dist):
    ans = 0.0
    for i in dist:
        ans += i * math.log(i, 2)
    return -ans
```

```python
entropy(distri)
```
